```python
high_rank_df = df[df['rank'] < 5]
```

```python
df['normalize_high_profit'] = df['high_profit'].apply(lambda x: 5 if x > 5 else x)
```

```python
df['normalize_high_profit'].groupby(pd.cut(df['rank'], np.arange(0, 101, 10))).mean()
```

```python
df['normalize_low_profit'] = df['low_profit'].apply(lambda x: -2 if x < -2 else x)
```

```python
df['normalize_low_profit'].groupby(pd.cut(df['rank'], np.arange(0, 101, 10))).mean()
```
